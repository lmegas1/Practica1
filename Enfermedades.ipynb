```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_json("enfermedades_mundo.json")
```

```python
# Se agrega Primera Columna: inicial del nombre de la enfermedad en mayúscula
data['inicial_enfermedad'] = data['nombre_enfermedad'].str[0].str.upper()
```

```python
data
```

```python
# Se agrega Segunda Columna: tasa de mortalidad (en porcentaje)
data['tasa_mortalidad'] = (data['muertes'] / data['numero_casos'] * 100).round(2)
```

```python
data
```

```python
# Se agrega Tercera Columna: década del año de reporte de enfermedad
data['decada'] = (data['anio'] // 10 * 10).astype(str) + 's'
```

```python
data
```

```python
# Se agrega Cuarta Columna: severidad clasificada según la tasa de mortalidad
def clasificar_severidad(tasa):
    if tasa < 2:
        return 'Leve'
    elif tasa < 6:
        return 'Moderada'
    elif tasa < 10:
        return 'Grave'
    else:
        return 'Muy Grave'
```

```python
data['severidad'] = data['tasa_mortalidad'].apply(clasificar_severidad)
```
