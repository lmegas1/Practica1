```python
#Práctica 1: Uso de aplicaciones colaborativas para el uso de herramientas de Inteligencia Artificial
#Definición del proyecto: Enfermedades a nivel mundial resuelto con las librerías/bibliotecas estudiadas.
#La información de entrada está en un archivo JSON

# Se crea el Notebook en la herramienta
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Se pasa los registros a un dataframe de pandas del conjunto de datos
data = pd.read_json("enfermedades_mundo.json")
```

```python
data.shape
```

```python
data.describe()
```

```python
data.keys()
```

```python
# Se agrega Primera Columna en función de los datos originales: "inicial del nombre de la enfermedad en mayúscula"
data['inicial_enfermedad'] = data['nombre_enfermedad'].str[0].str.upper()
```

```python
data
```

```python
# Se agrega Segunda Columna en función de los datos originales: "tasa de mortalidad (en porcentaje)"
data['tasa_mortalidad'] = (data['muertes'] / data['numero_casos'] * 100).round(2)
```

```python
data
```

```python
# Se agrega Tercera Columna en función de los datos originales: "década del año de reporte de enfermedad"
data['decada'] = (data['anio'] // 10 * 10).astype(str) + 's'
```

```python
data
```

```python
# Se agrega Cuarta Columna en función de los datos originales: "severidad clasificada según la tasa de mortalidad"
def clasificar_severidad(tasa):
    if tasa < 2:
        return 'Leve'
    elif tasa < 6:
        return 'Moderada'
    elif tasa < 10:
        return 'Grave'
    else:
        return 'Muy Grave'
```

```python
data['severidad'] = data['tasa_mortalidad'].apply(clasificar_severidad)
```

```python
data
```

```python
# Vista resumen de las cuatro columnas nuevas agregadas en función de los datos originales
data[['nombre_enfermedad', 'inicial_enfermedad', 'tasa_mortalidad', 'decada', 'severidad']].head(10)
```

```python
# Primera visualización del dataframe por defecto
data.groupby("pais").size().plot(kind="bar")
```

```python
# Segunda visualización del dataframe usando bokeh
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()
```

```python
conteo = data['pais'].value_counts().head(20)
```

```python
conteo
```

```python
p = figure(x_range=list(conteo.index),
           title="Los 20 países con mayor índice de enfermedades en el mundo",
           width=800, height=400)

p.vbar(x=list(conteo.index),
       top=conteo.values,
       width=0.6)

p.xaxis.major_label_orientation = 1
```

```python
show(p)
```

```python
# Otra visualización del dataframe usando matplotlib
# ── Top 10 enfermedades por casos totales
top10 = data.groupby("nombre_enfermedad")["numero_casos"].sum().nlargest(10)

plt.bar(top10.index, top10.values, color="steelblue")
plt.xticks(rotation=30, ha="right")
plt.title("Top 10 Enfermedades por Casos Totales")
plt.xlabel("Enfermedad")
plt.ylabel("Número de Casos (millones)")
plt.tight_layout()
plt.show()
```
